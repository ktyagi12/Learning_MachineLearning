```python
""" In case of skewed distribution, we will use IQR to calculate the boundaries.
# The new range will be [(Q1-IQR*1.5) - (Q3+IQR*1.5)]

# We need:
1. Q1, Q3, IQR

NB. If the outliers are at the extreme end, we will use IQR*3 instead of 1.5.
"""

Q1 = df_train['Annual_Premium'].quantile(0.25)
Q3 = df_train['Annual_Premium'].quantile(0.75)

IQR =  Q3 - Q1

#### Now the new range
lower_bridge = Q1 - (IQR*1.5)
upper_bridge = Q3 + (IQR*1.5)
print(lower_bridge, upper_bridge)
```

__*If a customer with Annual Premium value < lower_bridge or > upper_bridge is considered as outlier*__

## 2. Feature Engineering

1. Handle missing values
2. Remove rare categorical labels
3. Handling outliers
4. Encoding of categorical featues
5. Feature Transformation: Skewed to normal distribution
6. Handle imbalance dataset

### 2.1. Handling Missing Value

Since we don't have any missing values, skipping this section

### 2.2 Remove rare categorical labels

```python
for f in categorical_features:
    # Calculate the percentage of every category value
    perc = df_train.groupby(f)['Response'].count()/len(df_train)

    # Keep index of all those whose perc > 0.01
    temp = perc[perc > 0.01].index
    df_train[f] = np.where(df_train[f].isin(temp), df_train[f], 'Rare_Var')
```

### 2.3 Handling Outliers

```python
# For handling ouliers, we are replacing any value exceeding the boundary by upper bound and lower bound respectively
df_train.loc[df_train['Annual_Premium'] > upper_bridge, 'Annual_Premium'] = upper_bridge
df_train.loc[df_train['Annual_Premium'] < lower_bridge, 'Annual_Premium'] = lower_bridge
```

```python
# Histogram after removing outliers
fig = df_train['Annual_Premium'].hist(bins=50)
fig.set_xlabel('Annual_Premium')
```

```python
# Boxplot after removing outliers
df_train.boxplot(column='Annual_Premium')
```

### 2.4 Encoding

```python
df_train['Gender'] = pd.get_dummies(df_train['Gender'], drop_first= True)
df_train['Vehicle_Damage'] = pd.get_dummies(df_train['Vehicle_Damage'], drop_first= True)

map_dict = {'1-2 Year':1,'< 1 Year': 0, '> 2 Years':2}
df_train['Vehicle_Age'] =df_train['Vehicle_Age'].map(map_dict)
df_train.head()
```

```python
df_test['Gender'] = pd.get_dummies(df_test['Gender'], drop_first= True)
df_test['Vehicle_Damage'] = pd.get_dummies(df_test['Vehicle_Damage'], drop_first= True)

map_dict = {'1-2 Year':1,'< 1 Year': 0, '> 2 Years':2}
df_test['Vehicle_Age'] =df_test['Vehicle_Age'].map(map_dict)
df_test.head()
```

### 2.5 Feature Transfomation : Skewed to Normal distribution

__*We tried different transformation techniques (MinMax Scaler, Guassian transformation), there was no significant improvement in the data. Hence skipping this section. Also Classification model doesn't get affected by feature scaling.*__

### 2.6 Handling Imbalanced Dataset

```python
X_train = df_train.drop(['id', 'Response'], axis = 1)
y_train = df_train['Response']
```

```python
X_test = df_test.drop(['id', 'Response'], axis = 1)
y_test = df_test['Response']
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
smt = SMOTETomek(0.75)
X_train_smt, y_train_smt = smt.fit_resample(X_train, y_train)

print('The number of classes before fit:\n{}'.format(y_train.value_counts()))
print('\nThe number of classes after fit:\n{}'.format(y_train_smt.value_counts()))
```

## 3. Feature Selection

```python
# Apply Feature Selection
# First, we specify the Lasso Regression model, and then we select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then we use the selectFromModel object from sklearn, which will select the features which coefficients are non-zero.

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_train_smt, y_train_smt)
```

```python
feature_sel_model.get_support()
```

__True indicates feature is important, False the opposite.__

```python
# Let's see the total and the selected features
selected_features = X_train_smt.columns[(feature_sel_model.get_support())]

print('Total features are {}'.format(X_train_smt.shape[1]))
print('Selected features are {}'. format(len(selected_features)))
print('Features with coeff 0 are {}'.format(np.sum(feature_sel_model.estimator_.coef_ == 0))) # Skipped features, coeff becomes 0 by Lasso.
```

```python
X_train_smt = X_train_smt[selected_features]
X_train_smt.head()
```

## 4. Model Building

```python
clf = RandomForestClassifier()
clf.fit(X_train_smt, y_train_smt)
```

```python
X_test = df_test[selected_features]
```

```python
y_pred = clf.predict(X_test)
```
